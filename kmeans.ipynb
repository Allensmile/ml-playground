```python
from sklearn.datasets import make_blobs
import numpy as np

from unsupervised.kmeans import KMeans

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

import seaborn as sns
sns.set_context('notebook')
sns.set_style('white')

# for auto-reloading external modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
X, y = make_blobs(centers=4, n_samples=500, n_features=2,
                    shuffle=True, random_state=2046)
clusters = len(np.unique(y))

model = KMeans(K=clusters, max_iters=150)
model.fit(X)
```

```python
model.plot()
```

```python
from sklearn.cluster import KMeans as KMeans_SK

kmeans = KMeans_SK(n_clusters=clusters, max_iter=150).fit(X)
```

```python
for k in xrange(4):
    points = X[kmeans.labels_ == k].T
    plt.scatter(*points, c=sns.color_palette("hls", 4 + 1)[k])

for point in kmeans.cluster_centers_:
    plt.scatter(*point, marker='x', linewidths=10)
```
